```python
# 2번 문제

from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
driver = wb.Chrome()
driver.get("http://corners.gmarket.co.kr/Bestsellers")
time.sleep(1)
titleList = []
priceList = []
cateList = []
for i in range(10) :
    img = driver.find_elements_by_css_selector("img.lazy")
    img[i].click()
    time.sleep(1)
    title = driver.find_element_by_css_selector("h1.itemtit")
    titleList.append(title.text)
    price = driver.find_element_by_css_selector("strong.price_real")
    priceList.append(price.text)
    cate = driver.find_element_by_css_selector("li.on > a")
    cateList.append(cate.text)
    driver.back()
    time.sleep(1)

dic = {"이름" : titleList, "가격" : priceList, "카테고리" : cateList}
gmarket = pd.DataFrame(dic)
gmarket
```

```python
# 3번 문제

import requests as req
from bs4 import BeautifulSoup as bs

h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36"}
res = req.get("https://www.melon.com/chart/", headers = h)
soup = bs(res.text,"lxml")
song = soup.select("div.ellipsis.rank01 > span > a")
singer = soup.select("div.ellipsis.rank02 > span")
songList = []
singerList = []
rankList = []

for i in range(len(singer)) :
    songList.append(song[i].text)
    singerList.append(singer[i].text)
    rankList.append(i+1)
    
import pandas as pd
dic = {"노래제목" : songList, "가수" : singerList, "순위" : rankList}
melon = pd.DataFrame(dic)
melon.set_index("순위", inplace=True)
melon.to_csv("멜론차트.csv", encoding="euc_kr")
```

```python
# 4번 문제

from selenium import webdriver as wb
import time
import pandas as pd

driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(1)

try : 
    for i in range(30) :
        btn = driver.find_element_by_css_selector("a.c_05")
        btn.click()
        time.sleep(1)
        
except : 
    print("더보기 종료")
    
title = driver.find_elements_by_css_selector("h4.h.fz_03")
price = driver.find_elements_by_css_selector("div > div.item-text > div > strong")
titleList = []
priceList = []
    
for i in range(len(title)) :
    titleList.append(title[i].text)
    priceList.append(price[i].text)
dic = {"이름" : titleList, "가격" : priceList}
han = pd.DataFrame(dic)
han.to_csv("한솥.csv", encoding="euc_kr")
```
