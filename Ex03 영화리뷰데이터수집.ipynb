### 1페이지부터 10페이지까지 수집

```python
# 1페이지 정보를 수집하는 코드
for i in tq(range(1,101)):
    naver = req.get(f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=164122&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={i}")
    soup = bs(naver.text , "lxml")
    viewer = soup.select("span.ico_viewer")
    for i in viewer :
        i.extract()
    txt = soup.select("div.score_reple > p > span")
    for i in txt :
        print(i.text.strip())
```
