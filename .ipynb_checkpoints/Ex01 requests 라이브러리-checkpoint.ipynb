```python
import requests as req
```

# requests 
- 서버에게 데이터를 요청해서 응답받기 위한 라이브러리
- 요청하고, 응답받으면 역할이 끝
- 브라우저의 역할을 대신
- 페이지 정보를 알아오는 함수 .get("url")

```python
# req에게 네이버 메인페이지 정보 좀 알아오라고 명령
req.get("http://www.naver.com")
```

```python
# <Response [200]> : 데이터 통신에 성공했다.
# 400번대 에러 : 클라이언트(요청) 에러코드
# 500번대 에러 : 서버(응답) 에러코드
```

# Melon 사이트 페이지 정보를 요청

```python
# <Response [406]>
# 멜론서버에서 브라우저가 아닌 코드라고 인식했기 때문에 접근이 불가능
req.get("https://www.melon.com/")
```

```python
# 브라우저인척 속이는 방법
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.88 Safari/537.36"}
```

```python
 res = req.get("https://www.melon.com/", headers = h)
```

```python
# 응답객체(response)안에서 html 코드만 추출
res.text
```

# 네이버페이지에서 블로그,뉴스 글자 수집

```python
# 1. 네이버 메인페이지 정보를 요청
res = req.get("http://www.naver.com")
res.text
```

```python
# 3. 컴퓨터가 이해할 수 있는 html언어로 변형
# bs 객체화
# beautifulSoup 라이브러리 
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text, "lxml")
```

```python
# 2. 뉴스 블로그 등과 같은 글자만 수집
# 2.1 html 태그정보를 알려주기
# 필요한 요소를 검색해서 수집하는 명령 .select("선택자")★
soup.select("a.nav")
# 반환받은 리스트 데이터중에 첫번째 데이터만 가지고올때 select_one("선택자")★
soup.select_one("a.nav")
```

```python
# 반환받은 리스트 데이터중에 첫번째 데이터만 가지고올때 select_one("선택자")
soup.select_one("a.nav")
```
